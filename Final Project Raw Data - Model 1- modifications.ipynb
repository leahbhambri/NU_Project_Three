```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("final_project_data_raw.csv")
data
```

```python
data = data.drop(columns = "Unnamed: 0")
```

```python
data = data.drop(columns = "SERIALNO")
```

```python
data = data.drop(columns = "num_marriages")
```

```python
data["marital_status"].replace({2: 1, 4: 3}, inplace=True)
data
```

```python
data['marital_status'].value_counts()
```

```python
data = data.drop(columns = "state")
```

```python
data = data.dropna()
data
```

```python
# Key Question - what is something we want to know about?
# i.e. State - do we really want this factored
```

```python
data.iloc[:, :12]
```

```python
# Predetermined test vs train (most likely 75/25); random_state=1 makes the code reduceable
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:, :12], data['marital_status'], random_state=1)
```

```python
# StandardScaler removes the mean and scales each feature/variable to unit variance. 
# This operation is performed feature-wise in an independent way. 
# StandardScaler can be influenced by outliers (if they exist in the dataset) since it 
# involves the estimation of the empirical mean and standard deviation of each feature.
X_scaler = StandardScaler().fit(X_train)
```

```python
 X_test
```

```python
# Transform keeps the test data from the training data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_test_scaled[0]
```

```python
# one hot encoding translates the data to 0 False and 1 True
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

```python
y_test
```

```python
# 0 = false, 1 = true
# [row][column]
y_test_categorical[1]
```

```python
bool(0)
```

```python
# Create model and add layers
model = Sequential()
model.add(Dense(units=14, activation='relu', input_dim=12))
model.add(Dense(units=14, activation='relu'))

# needs 1 unit per column in one hot enconding
model.add(Dense(units=6, activation='softmax'))
```

```python
# what does this tell us?
model.summary()
```

```python
# Compile and fit the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=30,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```
