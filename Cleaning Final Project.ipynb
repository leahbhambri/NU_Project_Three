```python
import pandas as pd
import numpy as np
```

```python
#read in the files
ss13husa_raw = pd.read_csv("ss13husa.csv")
ss13husa_raw
```

```python
#select needed columnns
ss13husa = ss13husa_raw[['SERIALNO', 'ST', 'ACR','NRC',
                         'WORKSTAT', 'FES', 'FINCP', 'HUGCL', 'TAXP', 'MRGX',]]
ss13husa
```

```python
ss13husb_raw = pd.read_csv("ss13husb.csv")
ss13husb_raw
```

```python
#select needed columnns
ss13husb = ss13husb_raw[['SERIALNO', 'ST', 'ACR','NRC',
                         'WORKSTAT', 'FES', 'FINCP', 'HUGCL', 'TAXP', 'MRGX',]]
ss13husb
```

```python
#merge ss13husa and ss13husb to get ss13hus
ss13hus = ss13husa.append(ss13husb)
ss13hus
```

```python
ss13pusa_raw = pd.read_csv("ss13pusa.csv")
ss13pusa_raw
```

```python
#select needed columnns from ss13pusa
ss13pusa = ss13pusa_raw[['SERIALNO','MARHT', 'RAC1P', 'AGEP', 'SCHL','OCCP', 'MAR',]]
ss13pusa
```

```python
ss13pusb_raw = pd.read_csv("ss13pusb.csv")
ss13pusb_raw
```

```python
#select needed columnns from ss13pusb
ss13pusb = ss13pusb_raw[['SERIALNO','MARHT', 'RAC1P', 'AGEP', 'SCHL', 'OCCP', 'MAR']]
ss13pusb
```

```python
#merge ss13pusa and ss13pusb to get ss13pus
ss13pus = ss13pusa.append(ss13pusb)
ss13pus
```

```python
#merge ss13pus and ss13hus
merged_ph = pd.merge(ss13hus, ss13pus, on='SERIALNO', how='inner')
merged_ph
```

```python
#for column MAR, drop 2, 4 & 5
merged_ph_df = merged_ph[~merged_ph['MAR'].isin(['2', '4', '5'])]
merged_ph_df
```

```python
#rename columns
merged_ph_renamed = merged_ph.rename(columns = 
                                     {'ST': 'state',  
                                      'MAR': 'marital_status', 
                                      'MARHT': 'num_marriages', 
                                      'AGEP': 'age', 
                                      'ACR': 'lot_size', 
                                      'OCCP': 'job_type',  
                                      'NRC':'num_kids', 
                                      'SCHL': 'education', 
                                      'WORKSTAT':'work_status', 
                                      'FES':'fam_type_employment_status', 
                                      'FINCP':'family_income',   
                                      'HUGCL':'household_with_gp', 
                                      'RAC1P':'race', 
                                      'TAXP':'property_taxes', 
                                      'MRGX':'mortgage_status'})

merged_ph_renamed.tail(100)
```

```python
#for column state, convert codes to state names
merged_ph_renamed['state'] = merged_ph_renamed['state'].replace({1: 'AL', 2: 'AK', 3: '?', 4: 'AZ', 5: 'AR', 6: 'CA', 7: '?',
                                                               8: 'CO', 9: 'CT', 10: 'DE', 11: 'DC', 12: 'FL', 13: 'GA',
                                                               14: '?', 15: 'HI', 16: 'ID', 17: 'IL', 18: 'IN', 19: 'IA',
                                                               20: 'KS', 21: 'KY', 22: 'LA', 23: 'ME', 24: 'MD', 25: 'MA',
                                                               26: 'MI', 27: 'MN', 28: 'MS', 29: 'MO', 30: 'MT', 31: 'NE',
                                                               32: 'NV', 33: 'NH', 34: 'NJ', 35: 'NM', 36: 'NY', 37: 'NC',
                                                               38: 'ND', 39: 'OH', 40: 'OK', 41: 'OR', 42: 'PA', 43: '?',
                                                               44: 'RI', 45: 'SC', 46: 'SD', 47: 'TN', 48: 'TX', 49: 'UT',
                                                               50: 'VT', 51: 'VA', 52: '?', 53: 'WA', 54: 'WV', 55: 'WI',
                                                               56: 'WY', 72: "PR"})
merged_ph_renamed
```

```python
merged_ph_renamed['education'].replace({15: '[0] 12th grade - No Diploma', 
                                                               16: "[1] High School Diploma", 
                                                               17: '[2] GED',
                                                               18: '[3] Less Than 1 Year of College', 
                                                               19: "[4] Some College - No Degree",
                                                               20: "[5] Associate's", 
                                                               21: "[6] Bachelor's", 
                                                               22: "[7] Master's",
                                                               23: "[8] Professional Degree Beyond a Bachelor's", 
                                                               24: '[9] Doctorate Degree',
                                                               })
```

```python
#for column education, convert codes to TBD
merged_ph_renamed['education'] = merged_ph_renamed['education'].replace({15: '[0] 12th grade - No Diploma', 
                                                               16: "[1] High School Diploma", 
                                                               17: '[2] GED',
                                                               18: '[3] Less Than 1 Year of College', 
                                                               19: "[4] Some College - No Degree",
                                                               20: "[5] Associate's", 
                                                               21: "[6] Bachelor's", 
                                                               22: "[7] Master's",
                                                               23: "[8] Professional Degree Beyond a Bachelor's", 
                                                               24: '[9] Doctorate Degree',
                                                               })

merged_ph_renamed
```

```python
merged_ph_renamed['lot_size'] = merged_ph_renamed['lot_size'].fillna(0).replace({0 : "N/A", 
  1: ">1_acre", 
  2: '>10_acre', 
  3: '10+acres'})

merged_ph_renamed
```

```python
#convert race from numbers to strings
merged_ph_renamed['race'] = merged_ph_renamed['race'].replace({1: 'White',
                                                               2: "Black / African American",
                                                               3: 'American Indian',
                                                               4: 'Alaska Native',
                                                               5: "American Indian and Alaska Native",
                                                               6: "Asian",
                                                               7: "Native Hawaiian",
                                                               8: "Some Other Race",
                                                               9: "Two or More"
                                                               })
```

```python
merged_ph_renamed
```

```python
#change NaN to 0: num_marriages
merged_ph_renamed['num_marriages'] = merged_ph_renamed['num_marriages'].fillna(0)
```

```python
merged_ph_renamed
```

```python
#change NaN to none: job_type
merged_ph_renamed['job_type'] = merged_ph_renamed['job_type'].fillna(0).replace({0 : "none"})
```
